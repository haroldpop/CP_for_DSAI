```python
from cvxopt import matrix as cvxopt_matrix
from cvxopt import solvers as cvxopt_solvers
```

```python
from sklearn.model_selection import train_test_split
```

```python
class SVM:
  def __init__(self, X, y, kernel, C = None):
    self.X = X
    self.y = y
    self.C = C
    self.kernel = kernel
  
  def fit_data(self, test_size = None) :
    self.y[self.y == 0] = -1
    self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(self.X, self.y, test_size = test_size)

  def linear(self, x, z) :
    return np.dot(x, z.T)

  def polynomial(self, x, z, p=5):
    return (1 + np.dot(x, z.T)) ** p

  def gaussian(self, x, z, sigma=0.1):
     return np.exp(-np.linalg.norm(x - z, axis=1) ** 2 / (2 * (sigma ** 2)))

  def fit(self):
        X = self.X_train ##
        y = self.y_train ##
        m, n = X.shape
        # Calculate Kernel
        self.K = np.zeros((m, m))
        for i in range(m):
          for j in range(m) :
            if self.kernel == 'linear':
              self.K[i, j] = self.linear(X[i, np.newaxis], X[j, np.newaxis]) ##
            elif self.kernel == 'polynomial' :
              self.K[i,j] = self.polynomial(X[i, np.newaxis], X[j, np.newaxis])
            elif self.kernel == 'gaussian' :
              self.K[i,j] = self.gaussian(X[i, np.newaxis], X[j, np.newaxis])
            else:
              raise ValueError('Please select a linear, polynomial or gaussian kernel')
        # Solve with cvxopt final QP needs to be reformulated
        # to match the input form for cvxopt.solvers.qp
        #selection or not of the soft or hard margin method 
        P = cvxopt_matrix(np.outer(y, y) * self.K)
        q = cvxopt_matrix(-np.ones((m, 1)))
        A = cvxopt_matrix(y, (1, m))
        b = cvxopt_matrix(np.zeros(1))

        if self.C != None :
          G = cvxopt_matrix(np.vstack((- np.eye(m) , np.eye(m))))
          h = cvxopt_matrix(np.hstack((np.zeros(m), np.ones(m) * self.C)))

        else :
          G = cvxopt_matrix(-np.eye(m))
          h = cvxopt_matrix(np.zeros(m))
        
        #cvxopt_solvers.options["show_progress"] = False
        sol = cvxopt_solvers.qp(P, q, G, h, A, b)
        self.alphas = np.array(sol["x"])

  def predict(self):  #<----this is X_test
        y_predict = np.zeros((self.X_test.shape[0])) ##
        sv = self.get_parameters()

        for i in range(self.X_test.shape[0]): ##
          if self.kernel == 'linear' :
              y_predict[i] = np.sum(
                self.alphas[sv]
                * self.y_train[sv, np.newaxis]
                * self.linear(self.X_test[i], self.X_train[sv])[:, np.newaxis]
               )
          if self.kernel == 'polynomial' :
              y_predict[i] = np.sum(
                self.alphas[sv]
                * self.y_train[sv, np.newaxis]
                * self.polynomial(self.X_test[i], self.X_train[sv])[:, np.newaxis]
               )
          if self.kernel == 'gaussian' :
              y_predict[i] = np.sum(
                self.alphas[sv]
                * self.y_train[sv, np.newaxis]
                * self.gaussian(self.X_test[i], self.X_train[sv])[:, np.newaxis]
               )
        return np.sign(y_predict + self.b)

  def get_parameters(self):
        threshold = 1e-5
        if self.C != None :
            sv = ((self.alphas > threshold) * (self.alphas < self.C)).flatten()
        else :
            sv = (self.alphas > threshold).flatten()
        self.w = np.dot(self.X_train[sv].T, self.alphas[sv] * self.y_train[sv, np.newaxis])
        self.b = np.mean(
            self.y_train[sv, np.newaxis]
            - self.alphas[sv] * self.y_train[sv, np.newaxis] * self.K[sv, sv][:, np.newaxis]
        )
        return sv
```
